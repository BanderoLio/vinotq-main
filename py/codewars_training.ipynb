### **Kata**: Snail

(https://www.codewars.com/kata/521c2db8ddc89b9b7a0000c1)


```python
def snail(snail_map):
    if not snail_map or not snail_map[0]:
        return []

    n = len(snail_map)
    result = []

    up, down = 0, n - 1
    left, right = 0, n - 1

    while up <= down and left <= right:
        for i in range(left, right + 1):
            result.append(snail_map[up][i])
        up += 1

        for i in range(up, down + 1):
            result.append(snail_map[i][right])
        right -= 1

        if up <= down:
            for i in range(right, left - 1, -1):
                result.append(snail_map[down][i])
            down -= 1

        if left <= right:
            for i in range(down, up - 1, -1):
                result.append(snail_map[i][left])
            left += 1

    return result
```

### **Kata**: Multiply to `n`

(https://www.codewars.com/kata/5f1891d30970800010626843/train/python)


```python
def multiply(n, k):
    dp = [[0] * (k + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        dp[i][1] = 1

    for i in range(1, n + 1):
        for j in range(2, k + 1):
            for x in range(1, i + 1):
                if i % x == 0:
                    dp[i][j] += dp[i // x][j - 1]

    return dp[n][k]


multiply(24, 2)
# timeout :(
```

### **Kata**: VigenÃ¨re Cipher Helper

(https://www.codewars.com/kata/52d1bd3694d26f8d6e0000d3)


```python
class VigenereCipher(object):
    def __init__(self, key, alphabet):
        self.key = key
        self.alphabet = alphabet + alphabet

    def encode(self, text):
        ext_key = (self.key * ((len(text) // len(self.key)) + 1))[: len(text)]
        encoded_text = ""

        for i in range(len(text)):
            if text[i] in self.alphabet:
                encoded_text += self.alphabet[
                    (self.alphabet.index(text[i]) + self.alphabet.index(ext_key[i]))
                ]
            else:
                encoded_text += text[i]

        return encoded_text

    def decode(self, text):
        ext_key = (self.key * ((len(text) // len(self.key)) + 1))[: len(text)]
        decoded_text = ""

        for i in range(len(text)):
            if text[i] in self.alphabet:
                decoded_text += self.alphabet[
                    (self.alphabet.index(text[i]) - self.alphabet.index(ext_key[i]))
                ]
            else:
                decoded_text += text[i]

        return decoded_text


abc = "abcdefghijklmnopqrstuvwxyz"
key = "password"
c = VigenereCipher(key, abc)
print(c.encode("codewars"))
```

### **Kata**: ROT13

(https://www.codewars.com/kata/52223df9e8f98c7aa7000062/)


```python
def rot13(message):
    result = ""
    for letter in message:
        if letter.isalpha():
            if 65 <= ord(letter) <= 90:
                if 65 <= ord(letter) + 13 <= 90:
                    result += chr(ord(letter) + 13)
                else:
                    result += chr(ord(letter) - 13)

            elif 97 <= ord(letter) <= 122:
                if 97 <= ord(letter) + 13 <= 122:
                    result += chr(ord(letter) + 13)
                else:
                    result += chr(ord(letter) - 13)
            else:
                result += letter
        else:
            result += letter

    return result


print(rot13("EBG13 rknzcyr."))
print(
    rot13(
        "How can you tell an extrovert from an\nintrovert at NSA? Va gur ryringbef,\ngur rkgebireg ybbxf ng gur BGURE thl'f fubrf."
    )
)
print(rot13("123"))
print(rot13("Guvf vf npghnyyl gur svefg xngn V rire znqr. Gunaxf sbe svavfuvat vg! :)"))
print(rot13("@[`{"))
```

```python
import math


def is_sqr(x):
    return int(math.sqrt(x)) ** 2 == x


def sum_of_squares(n):
    if is_sqr(n):
        return 1

    for i in range(1, int(math.sqrt(n)) + 1):
        if is_sqr(n - (i * i)):
            return 2

    while n % 4 == 0:
        n >>= 2

    if n % 8 == 7:
        return 4

    return 3


print(sum_of_squares(15))
```

```python
# You are given a car odometer which displays the miles traveled as an integer. The odometer has a defect, however: it proceeds from digit 3 to digit 5 always skipping the digit 4. This defect shows up in all positions (ones, tens, hundreds, etc).
def faulty_odometer(n):
    ls_of_nms = {"0": 0, "1": 1, "2": 2, "3": 3, "5": 4, "6": 5, "7": 6, "8": 7, "9": 8}
    str_n = str(n)
    res = 0

    for i, num in enumerate(str_n):
        res = res * 9 + ls_of_nms[num]

    return res
    # print(res)


faulty_odometer(13)
faulty_odometer(15)
faulty_odometer(55)
faulty_odometer(2003)
```

```python
import re

def normalize_equation(coeffs, rhs):
    if not coeffs:
        return tuple()
    
    max_coeff = max(abs(v) for v in coeffs.values())

    if max_coeff == 0:
        return tuple()
    
    normalized = [(var, coeff / max_coeff) for var, coeff in sorted(coeffs.items())]
    normalized.append(('_rhs', rhs / max_coeff))
    
    return tuple(normalized)

def parse_equations(equations):
    variables = set()
    parsed_equations = []
    unique_equations = set()

    for eq in equations:
        left, right = [side.strip() for side in eq.split("=")]
        coeffs = {}

        def parse_terms(expr: str, multiply: float = 1.0) -> float:
            constant = 0.0
            terms = re.findall(r"([+-]?\d*\.?\d*)?([a-zA-Z][a-zA-Z0-9_]*)?", expr)

            for coeff_str, var in terms:
                if not coeff_str and not var:
                    continue

                if var:
                    coeff = 1.0
                    if coeff_str:
                        if coeff_str in ["+", ""]:
                            coeff = 1.0
                        elif coeff_str == "-":
                            coeff = -1.0
                        else:
                            coeff = float(coeff_str)
                    coeffs[var] = coeffs.get(var, 0.0) + coeff * multiply
                    variables.add(var)
                elif coeff_str:
                    try:
                        constant += float(coeff_str)
                    except ValueError:
                        continue

            return constant

        const_left = parse_terms(left)
        const_right = parse_terms(right, -1.0)
        rhs = const_right - const_left

        normalized = normalize_equation(coeffs, rhs)
        if normalized and normalized not in unique_equations:
            unique_equations.add(normalized)
            parsed_equations.append((coeffs, rhs))

    if not variables:
        return None, None, None

    var_list = sorted(variables)
    var_index = {var: i for i, var in enumerate(var_list)}

    m = len(var_list)  
    n = len(parsed_equations) 

    if n < m:
        return None, None, None

    if n > m:
        parsed_equations = parsed_equations[:m]
        n = m

    A = [[0.0] * m for _ in range(n)]
    B = [0.0] * n

    for i, (coeffs, rhs) in enumerate(parsed_equations):
        for var, coeff in coeffs.items():
            A[i][var_index[var]] = coeff
        B[i] = rhs

    return A, B, var_list

def check_system_consistency(A, B):
    n = len(A)
    for i in range(n):
        if all(abs(A[i][j]) < 1e-10 for j in range(n)) and abs(B[i]) > 1e-10:
            return False
    return True

def determinant(matrix):
    n = len(matrix)

    if n == 0:
        return 0
    if n == 1:
        return matrix[0][0]
    if n == 2:
        return matrix[0][0] * matrix[1][1] - matrix[0][1] * matrix[1][0]

    det = 0.0
    for j in range(n):
        minor = [row[:j] + row[j + 1:] for row in matrix[1:]]
        det += (-1) ** j * matrix[0][j] * determinant(minor)

    return det

def cramer(A, B):
    n = len(A)
    det_A = determinant(A)

    if not det_A:
        if not check_system_consistency(A, B):
            return None  
        return None 

    solutions = []
    for i in range(n):
        Ai = [row[:] for row in A]
        for j in range(n):
            Ai[j][i] = B[j]
        solutions.append(determinant(Ai) / det_A)

    return solutions

def solve(*equations):
    A, B, variables = parse_equations(equations)
    
    if A is None or B is None or variables is None:
        return None
        
    solutions = cramer(A, B)
    
    if solutions is None:
        return None
        
    return {var: sol for var, sol in zip(variables, solutions)}
```
