### **Lection #2**
#### **Date: 17.02.25**
#### **Theme**: **Связывание** (*линковка*)

#### **Линковка** - процесс сборки всех частей программмного кода в единый запускаемый файл или библиотеку. Один из последних этапов компиляции.

## **Процесс компиляции**:
1. **Предпроцессинг** - `&gcc -E` - как результат - преобразование исходного кода с макроподстановками
2. **Трансляция** - `&gcc -S` - перевод в коды ассемблера
3. **Ассемблирование** - `&gcc -с (ar)` - перевод в бинарники
4. **Линковка**:   
    a. `&gcc -static (ld)` - ***статическое связывание***,  появление ссылок на реальные функции, круто в многофайловом проекте так как не нужно будет перекомпайлить весь код: можно перекомпайлить только изменившуюся часть  
    b. `&gcc (ld)` - ***динамическое связывание***, аналогично, но в своем случае позволяет использовать библиотеки единожды, а не создавать их для каждой программы новую
